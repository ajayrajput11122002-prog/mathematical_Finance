```python
"""
Monte Carlo simulation for random variable expectation
"""

import numpy as np


def simulate_random_variable(n_samples):
    """
    Simulate X ~ N(0,1)

    Parameters
    ----------
    n_samples : int
        Number of Monte Carlo samples

    Returns
    -------
    samples : np.ndarray
    """
    samples = np.random.randn(n_samples)
    return samples


def monte_carlo_expectation(samples):
    """
    Estimate E[X] using sample mean
    """
    return samples.mean()


def monte_carlo_variance(samples):
    """
    Estimate Var(X) using sample variance
    """
    return samples.var()
```
